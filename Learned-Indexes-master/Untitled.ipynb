```python
import pandas as pd
data = pd.read_csv('data/exponential.csv', header=None)

exponential = []
label = []
for i in range(1000):
    label.append(i)
    exponential.append(data.iloc[i, 0])
    
import numpy as np
import matplotlib.pyplot as plt 
x=np.arange(0,len(label))
l1=plt.plot(label,exponential,'r--',label='exponential')
plt.title('The Exponential Data')
plt.xlabel('row')
plt.ylabel('column')
plt.legend()
plt
```

```python
def train(inputs,labels):
    n = len(inputs)  # 获取x的长度，x是list
    dinominator = 0  # 初始化分母
    numerator = 0  # 初始化分子
    for i in range(0, n):  # 求b1
        numerator += (inputs[i] - np.mean(inputs)) * (labels[i] - np.mean(labels))
        dinominator += (inputs[i] - np.mean(inputs)) ** 2  # **表示平方

    print("numerator:" + str(numerator))
    print("dinominator:" + str(dinominator))

    b1 = numerator / float(dinominator)  # 得出b1
    b0 = np.mean(labels) - b1 * float(np.mean(inputs))  # 得出b0
    
    error = 0
    for i in range(n):
        y_predict = b1 * inputs[i] + b0
        error +=  abs(y_predict - labels[i])
    mean_error = error / n * 1.0
    return b1,b0,mean_error
```

```python
w,b,me = train(linear,label)
```
